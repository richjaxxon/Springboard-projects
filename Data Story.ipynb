# A Bird of a Different Feather

### Are slight physical differences in attributes sufficient to determine species classification?

We shall look at this data set and answer the question, "with such minute differences, can we determine what species they are?". 

```python
# import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
# get data 

penguins = pd.read_csv('https://raw.githubusercontent.com/mcnakhaee/palmerpenguins/master/palmerpenguins/data/penguins.csv')
penguins.head()
```

# Correlations

How do the physical characteristics correlate 

```python
corr = penguins.corr()
corr.style.background_gradient(cmap='coolwarm')
```

## Relationships

Let's look at the relationships between attributes and species

```python
g = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=7,
               data=penguins,
               palette=['#FF8C00','#159090','#A034F0'])
g.set_xlabels('Flipper Length')
g.set_ylabels('Body Mass')
```

## Normal Distributions

The data seems to follow a normal distribution of attributes among the birds. This is promising.

```python
sns.displot(
    penguins, x="flipper_length_mm", col="species", row="sex",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

## See how the penguins differ by sex

```python
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```

## See how the penguins differ by location 

Does a different location have any influence on physical traits? 

```python
plt.figure(figsize = [10,7])
g = sns.boxplot(x = 'island',
            y ='body_mass_g',
            hue = 'species',
            data = penguins,
            palette=['#FF8C00','#159090','#A034F0'],
            linewidth=0.3)
g.set_xlabel('Island')
g.set_ylabel('Body Mass')
```

# Conclusion

It would appear that regardless of sex, location, or similarity of attributes, we can indeed classify penguins by species with the ratios of physical traits.
